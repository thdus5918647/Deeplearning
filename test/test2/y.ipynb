```python
y = []    # 빈 리스트 생성
 
for x in range(10):
    line = []              # 안쪽 리스트로 사용할 빈 리스트 생성
    for j in range(1311):
        line.append(0)     # 안쪽 리스트에 0 추가
    y.append(line)         # 전체 리스트에 안쪽 리스트를 추가
 
print(y)
```

```python
print(len(y))
print(len(y[0]))
```
