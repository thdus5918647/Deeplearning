### 1. basic data

```python
import pandas as pd

path_file = pd.read_csv('./basic.csv')
path_file.head()
```

```python
path_file['y'] = 0
path_file.head()
```

```python
path_file.tail()
```

```python
bins = list(range(0,7501,50))
```

```python
bins_label = [int(x) for x in bins]
print(bins_label)
```

```python
path_file["hz_level"] = pd.cut(path_file['hz'], bins, right=False, labels=bins_label[:-1]) 
```

```python
path_file
```

```python
path_file.hz_level.value_counts().to_frame().style.background_gradient(cmap='summer')
```

```python
basic = path_file.loc[0:999]
basic.tail()
```

```python
basic.to_csv("basic_data.csv", header=True, index=False)
len(basic)
```
